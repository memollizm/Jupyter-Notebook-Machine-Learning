```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,3,5,7,11]
plt.plot(x,y)
plt.xlabel('X ekseni')
plt.ylabel('Y ekseni')
plt.title('Çizgi Grafiği')
plt.show()
```

```python
veri = [1,1,2,3,3,3,4,4,5,5,5,5,5]

plt.hist(veri, bins=5) #bins keyword ne işe yarıyor
plt.xlabel('Değerler')
plt.ylabel('Frekans')
plt.title('Histogram')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,3,5,7,11]
plt.scatter(x,y)
plt.xlabel('X ekseni')
plt.ylabel('Y ekseni')
plt.title('Dağılım Grafiği')
plt.show()
```

```python
import seaborn as sns
tips = sns.load_dataset('tips')
sns.boxplot(x='day', y='total_bill', data=tips, palette='Set3')
plt.xlabel('Gün')
plt.ylabel('Toplam Hesap')
plt.title('Gün ve Toplam Hesap İlişkisi')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
corr = sns.load_dataset('tips').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Korelasyon Matrisi Sıcaklık Haritası')
#plt.show()
```

```python
import seaborn as sns
import pandas as pd
veri = pd.DataFrame({'A': [1,2,3,4,5],
                     'B': [2,4,1,5,2],
                     'C': [5,3,2,1,4]})
sns.pairplot(veri)
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
g = sns.FacetGrid(titanic, col="class", row="sex")
g.map(plt.hist, "age")
plt.show()
```

```python
tips = sns.load_dataset('tips')
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
plt.show()
```
