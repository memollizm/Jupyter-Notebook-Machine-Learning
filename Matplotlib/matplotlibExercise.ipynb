```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

### Example 1

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

### Example 2

```python
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, .2, .2])

ax1.plot(x,y, color='red')
ax1.set_xlabel=('X')
ax1.set_ylabel=('y')

ax2.plot(x,y, color='red')
ax2.set_xlabel('X')
ax2.set_ylabel('y')
```

### Example 3

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2, 0.5, .4, .4])

axes1.plot(x,z)
axes1.set_xlabel('X')
axes1.set_ylabel('Z')

axes2.plot(x,y)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
```
