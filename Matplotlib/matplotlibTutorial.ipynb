```python
import matplotlib.pyplot as plt
```

```python
#Grafiklerin tarayıcıda ayrı bir pencerede değil, doğrudan notebook hücresinin altında görüntülenmesi için
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x **2
```

```python
print("x: ", x)
print("y: ", y)
```

## Functional

```python
plt.plot(x,y)
plt.xlabel('X Ekseni')
plt.ylabel('Y Ekseni')
plt.title('Başlık')
```

```python
#Birden çok alt grafik oluşturmayı sağlar.
#plt.subplot(satır, sütun, index)
plt.subplot(1,2,1)
plt.plot(x,y, 'b')

plt.subplot(1,2,2)
plt.plot(y,x, 'r')
```

## Object Oriented

* Grafiklerin daha ayrıntılı olarak özelleştirilmesi gerektiğinde kullanışlıdır.

```python
fig = plt.figure()

#Başlangıç noktası (0.1, 0.1) olan dikdörtgenin çizilmesi
#Genişlik ve yüksekliği sırasıyla (0.8, 0.8) olduğunun belirtilmesi
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x,y)
axes.set_xlabel('X ekseni')
axes.set_ylabel('Y ekseni')
axes.set_title('Başlık')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2= fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('İlk Çizim')

axes[1].plot(y,x)
axes[1].set_title('İkinci Çizim')

#Alt grafiklerdeki bindirmeleri ve çakışmaları gidermek için
plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X kare')
ax.plot(x,x**3, label='X küp')

#Grafikteki çizgilerin ne anlama geldiğini göstermek için.
#loc=0 parametresi grafiğe göre açıklama kutucuğunu en iyi konuma koyar.
ax.legend(loc=0)
```

## Plot Görünümleri

* <b>`ax.plot(x,y, color='', lw=, alpha=, ls='', marker='')` </b>
    * `color=''` çizgi rengi için kullanılır [HEX değerleri alabilir]
    * `lw=` çizgi genişliği için kullanılır
    * `alpha=` çizginin şeffaflığı için kullanılır
    * `ls=''` çizginin stili için kullanılır
    * `marker=''` X üzerindeki her noktayı işaretleyen işaretçi oluşturur.

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='green', lw=3, ls='-', marker='o')
```
