```python
# Gerekli kütüphaneleri yükleyelim
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Örnek veri seti oluşturalım
X = np.array([1, 2, 3, 4, 5])  # Bağımsız değişken
y = np.array([5, 8, 11, 19, 26])  # Bağımlı değişken
X
```

```python
# Veri setini görselleştirelim
plt.scatter(X, y, color='blue')
plt.xlabel('Bağımsız Değişken (X)')
plt.ylabel('Bağımlı Değişken (y)')
plt.title('Basit Lineer Regresyon Örneği - Veri Seti')
plt.show()
```

```python
# Veri setini görselleştirelim
plt.plot(X, y, color='blue')
plt.xlabel('Bağımsız Değişken (X)')
plt.ylabel('Bağımlı Değişken (y)')
plt.title('Basit Lineer Regresyon Örneği - Veri Seti')
plt.show()
```

```python
# X'i ve y'yi uygun şekilde yeniden şekillendirerek X matrisini oluşturalım
X = X.reshape(-1, 1)
```

```python
X
```

```python
# Basit Lineer Regresyon modelini oluşturalım
from sklearn.linear_model import LinearRegression

model = LinearRegression()

# Modeli eğitelim
model.fit(X, y)
```

```python
# Eğitilmiş modeli kullanarak tahminler yapalım
y_pred = model.predict(X)
```

```python
# Eğitilmiş modelin doğrusunu ve veri setini görselleştirelim
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')  # Tahmin edilen doğru
plt.xlabel('Bağımsız Değişken (X)')
plt.ylabel('Bağımlı Değişken (y)')
plt.title('Basit Lineer Regresyon - Model Doğrusu')
plt.show()
```

```python
# Modelin eğim katsayısını ve kesme terimini yazdıralım
print("Eğim (Katsayı):", model.coef_[0])
print("Kesme Terimi:", model.intercept_)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
# R-kare (R-squared) hesaplama
r2 = r2_score(y, y_pred)
print("R-kare (R-squared):", r2)

# MSE hesaplama
mse = mean_squared_error(y, y_pred)
print("Ortalama Kare Hata (MSE):", mse)

# RMSE hesaplama
rmse = np.sqrt(mse)
print("Kök Ortalama Kare Hata (RMSE):", rmse)
```

```python
#(0.9528) oldukça yüksek bir R-kare değeri olduğundan, modelin bağımsız değişkenlerin büyük bir kısmının bağımlı değişken üzerindeki varyansı açıkladığı söylenebilir. 
#Yani, modelin verilere iyi uyum sağladığı yorumu yapılabilir.

#Verilen MSE değeri (2.78) ve düşük MSE değeri daha iyi bir model performansını gösterir. 
#Bu durumda, elde edilen MSE değeri oldukça düşük olduğu için, modelin tahminlerinin gerçek değerlerden ortalama olarak ne kadar uzak olduğunun düşük olduğu söylenebilir.

#RMSE değeri (1.6673), modelin tahminlerinin gerçek değerlere olan ortalama uzaklığını temsil eder. Daha düşük RMSE değeri, daha iyi bir model performansını gösterir. 
#Verilen RMSE değeri oldukça düşük olduğundan, modelin gerçek değerlerle olan ortalama uzaklığının düşük olduğu söylenebilir.
```
